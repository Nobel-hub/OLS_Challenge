```python
from scipy.stats import normaltest
numeric_column = new_df.select_dtypes(include=np.number).columns
gaussian_column = []
non_gaussian_column = []
for column in numeric_column:
    stats, p = normaltest(new_df[column])
    alpha = 0.05
    if p>alpha:
        gaussian_column.append(column)
    else:
        non_gaussian_column.append(column)
gaussian_column
```

```python
plt.hist(new_df['PctPublicCoverage'], bins=30)
plt.show()
```

```python
len(non_gaussian_column)
```

```python
gaussian_data = new_df[gaussian_column]
basic_info_gaussian = gaussian_data.describe()
basic_info_gaussian = basic_info_gaussian.loc[['mean', 'std', 'min', 'max']]
basic_info_gaussian
```

```python
print("Highest allowed: ", basic_info_gaussian["PctPublicCoverage"]["mean"] + 3 * basic_info_gaussian["PctPublicCoverage"]["std"])
print("Lowest allowed: ", basic_info_gaussian["PctPublicCoverage"]["mean"] - 3*basic_info_gaussian["PctPublicCoverage"]["std"])
outliers = new_df[(new_df['PctPublicCoverage'] > 59.777864755203126) | (new_df['PctPublicCoverage'] < 12.72741913058617)]
outliers['PctPublicCoverage'].head()
```

```python
def deal_with_outliers(df, col, basic_info):
    highest = basic_info[col]["mean"] + 3*basic_info[col]["std"]
    lowest = basic_info[col]["mean"] - 3*basic_info[col]["std"]
    df = df[(df[col]>highest) | (df[col]<lowest)]
    return df
cols_have_outliers = []
for col in gaussian_column:
    df = deal_with_outliers(gaussian_data, col, basic_info_gaussian)
    shape = df.shape
    if shape[0] > 0:
        cols_have_outliers.append(col)
print(cols_have_outliers)
```

```python
for cols in cols_have_outliers:
    highest = basic_info_gaussian[col]["mean"] + 3*basic_info_gaussian[col]["std"]
    lowest = basic_info_gaussian[col]["mean"] - 3*basic_info_gaussian[col]["std"]
    trimmed_data = new_df[(new_df[cols] < highest) & (new_df[cols]>lowest)]
trimmed_data
```

```python
for cols in cols_have_outliers:
    highest = basic_info_gaussian[col]["mean"] + 3*basic_info_gaussian[col]["std"]
    lowest = basic_info_gaussian[col]["mean"] - 3*basic_info_gaussian[col]["std"]
    capped = new_df.copy()
    capped.loc[capped[cols]>highest, cols] = highest
    capped.loc[capped[cols]<lowest, cols] = lowest
capped
```

```python
cols_to_drop = []
for col in new_df.columns:
    if len(new_df[col].unique()) < 10:
        cols_to_drop.append(col)
print(len(cols_to_drop))
data_for_skewness = new_df.drop(cols_to_drop, axis=1)
```

```python
def identify_skewed_cols(df, cols):
    skewed_cols = []
    for col in cols:
        if df[col].skew() > 1 or df[col].skew() < -1:
            skewed_cols.append(col)
    return skewed_cols
skewed_cols = identify_skewed_cols(data_for_skewness, data_for_skewness.columns)
len(skewed_cols)
```

```python
skewed_data = data_for_skewness[skewed_cols]
skewed_data.shape
```

```python
fig, axes = plt.subplots(nrows = 4,ncols=4, figsize=(20,10))
for i, col in enumerate(skewed_cols):
    axes[i//4, i%4].set_title(col)
    skewed_data[col].plot(kind='density', ax =axes[i//4, i%4], subplots=True, sharex=False )
plt.tight_layout()
plt.show()
```

```python
def Box_plot(df, col):
    figure = go.Figure(data=[go.Box(y=df[col], boxpoints='outliers', jitter = 0.3, pointpos=1.8)])
    figure.update_layout(
        title="Boxplot for "+ col,
        yaxis_title = col,
        height = 500,
        width=500
    )
    figure.show()
for col in skewed_cols:
    Box_plot(new_df, col)
```

```python
q1 = new_df['studyPerCap'].quantile(0.25)
q3 = new_df['studyPerCap'].quantile(0.75)
iqr = q3-q1
upper_limit = q3 + 1.5 * iqr
lower_limit = q1-1.5*iqr
new_df[(new_df['studyPerCap']>upper_limit) | new_df['studyPerCap']<lower_limit]
print(upper_limit, lower_limit)
```

```python
def deal_outlier_with_iqr(df, cols):
    q1 = df[cols].quantile(0.25)
    q3 = df[cols].quantile(0.75)
    iqr = q3-q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1-1.5*iqr
    outliers = df[(df[cols]>upper_limit) | (df[cols]<lower_limit)]
    return outliers
cols_have_outliers_skewed = []
for cols in skewed_cols:
    outliers = deal_outlier_with_iqr(new_df, cols)
    shape = outliers.shape
    if shape[0]>0:
        cols_have_outliers_skewed.append(cols)
cols_have_outliers_skewed
```

```python
def capping(df, col):
    q1 = df[cols].quantile(0.25)
    q3 = df[cols].quantile(0.75)
    iqr = q3-q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1-1.5*iqr
    df.loc[df[col] > upper_limit, col] = upper_limit
    df.loc[df[col] < lower_limit, col] = lower_limit
    return df
for col in cols_have_outliers_skewed:
    capped_data = capping(new_df, col)
capped_data.shape
```
